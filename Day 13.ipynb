```python
# List of GK Questions
import time
import pygame
import IPython
pygame.init()


sound_path = r"C:\Users\parmeet\Downloads\kbc_5_newest.mp3"
theme_sound = pygame.mixer.Sound(sound_path)
theme_sound.play()
print('Welcome to KBC'.upper().center(100,'*'))

cr_7_sound_path = r"C:\Users\parmeet\Downloads\7cr.mp3"

questions = [
    "What is the capital of France?",
    "Who wrote the play 'Romeo and Juliet'?",
    "Which planet is known as the Red Planet?",
    "In which year did World War II end?",
    "What is the largest ocean on Earth?"
]

# List of MCQ Options for each question
options = [
    ["Paris", "Berlin", "Madrid", "London"],
    ["William Shakespeare", "Jane Austen", "Charles Dickens", "George Orwell"],
    ["Mars", "Venus", "Jupiter", "Saturn"],
    ["1940", "1942", "1945", "1950"],
    ["Atlantic Ocean", "Indian Ocean", "Arctic Ocean", "Pacific Ocean"]
]

# List of Correct Answers for each question (corresponding to the above options)
correct_answers = ["a","a","a","c","d"]

op = 'abcd'

each_ans = []
win_amount = 0

for i,j in enumerate(questions):
    print(f'Question No {i+1}. {j}')
    
    for k,l in enumerate(options[i]):
        print(f'''    {op[k]}.  {l}''')


    user_ans = input('👉Enter your Choice: ').lower()
    if (user_ans in 'abcd') or (user_ans in ['exit','quit','e','q','x']):
        if user_ans in 'abcd':
            sahi_ans = correct_answers[i]
            if user_ans == sahi_ans:
                print('👌Great Reply, Ready for next question!!',end = '')
                each_ans.append(True)
                
                for s in range(5):
                    print('..', end = '')
                    time.sleep(1)
    
                print(end = '\n'*3)
                
                if win_amount == 0:
                    win_amount += 10_000
                    print('🤑🤑 10K win!!')
                    
                    sound_path = r"C:\Users\parmeet\Downloads\spinning-coin-on-table-352448.mp3"
                    coin_sound = pygame.mixer.Sound(sound_path)
                    coin_sound.play()    
                    
                else:
                    print(f'🤑🤑You Won Rs {win_amount*2}')
                    win_amount = win_amount + win_amount*2

                    sound_path = r"C:\Users\parmeet\Downloads\spinning-coin-on-table-352448.mp3"
                    coin_sound = pygame.mixer.Sound(sound_path)
                    coin_sound.play()  
            else:
                print('😒😒You have Lost this game!!')   
                break
        else:
            print('😊😊Thanks for Playing😊😊')
            break
    else:
        print('❌❌Invalid Selection!!')
        break
           
    print()
    # display(clear = True)

print(f'🤑🤑Total Win Amount: Rs {win_amount}')

if len(each_ans) == len(questions):
    pygame.mixer.Sound(cr_7_sound_path).play()
    display(IPython.display.Image('https://i.makeagif.com/media/4-23-2025/xtKC3c.gif'))
```
