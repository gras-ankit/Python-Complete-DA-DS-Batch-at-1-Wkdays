# For Loop Pass, Break, Continue, Nested For Loop, For Loop With Else

```python
passengers = {"Gyan":{'Amount':450},'Riya':{'Amount':50},'Sakib':{'Amount':78},
          'Harsh':{'Amount':-10},
          'Ekansh':{'Amount':1000}}

print(passengers)
```

```python
import IPython
import time

for i in passengers:
    car_balance = passengers[i]['Amount']
    if car_balance >= 10:
        url = 'https://media.tenor.com/MLG6yk-hRVsAAAAM/train-doors.gif'

        print(f'{i}, you are allowed')
        display(IPython.display.Image(url, width = 100))
    else:
        url = 'https://media.tenor.com/xrgfudiAFW8AAAAM/missed-train-late.gif'
        display(IPython.display.Image(url,width=100))
        print(f'{i}, you are not allowed')

    display(clear= True)
        
    time.sleep(3)
display(clear= True)
print('All Done')
```

# Break

```python
passengers = {"Gyan":{'Luggage':['Books','Laptop','Wallet']},
              'Riya':{'Luggage':['Books','Bottle','Purse']},
              'Sakib':{'Luggage':['Resume','Passport','Bag']},
          'Harsh':{'Luggage':['Books','Knife','Resume']},
          'Ekansh':{'Luggage':[None]},
          'Dhruv':{'Luggage':['Books','Laptop']}}

print(passengers)
```

```python
banned_items = [
    "Knife",
    "Firearms",
    "Explosives",
    "Sharp objects",
    "Illegal drugs",
    "Flammable substances",
    "Acids or corrosive chemicals",
    "Gas canisters",
    "Toxic materials",
    "Large or bulky luggage",
    "Alcohol (in some regions)",
    "Pets (unless in carrier)",
    "Bicycles (in some regions)",
    "Skateboards",
    "Glass bottles",
    "Ammunition",
    "Fireworks",
    "Pepper spray",
    "Self-defense weapons (such as tasers)",
    "Loudspeakers or amplifiers"
]
print(banned_items)
```

```python
import winsound
import IPython
import time

result = False

for name in passengers:
    passenger_luggage = passengers[name]['Luggage']
    for luggage in passenger_luggage:
        if luggage not in banned_items:
            pass
        else:
            print(f'{name} Your Luggage contained Banned Item, you are not allowed')
            
            url = 'https://media.tenor.com/vKFl6YLvErMAAAAM/cops-running-from-tackle.gif'
            display(IPython.display.Image(url,width=100))
            winsound.Beep(2000,5000)
            result  = True
            break   # break< it will completely break the code and end loop
    
    if result == True:
        break
    else:
        print(f'{name}, you can go!!')
        # Play a WAV file asynchronously
        winsound.PlaySound(r"C:\Users\salma\Downloads\success-340660.mp3", winsound.SND_FILENAME | winsound.SND_ASYNC)
    time.sleep(3)
    display(clear=True)
```
