# Number Guess Game

```python
import random
import time
print('Check your Luck!!'.center(100,'*'))

comp_choice = random.randint(1,10)
user_choice  = int(input('Enter your Choice (1-10): '))


print('Wait!!')
time.sleep(3)

print(f'''
Computer choice: {comp_choice}
User choice: {user_choice}
''')

if user_choice == comp_choice:
    print('You Won!!💰')
else:
    print('Better Luck Next Time!!🥹😥')
```

# Rock Paper Scissors Game

```python
import random
choices = ['rock','paper','scissors']
comp_choice = random.choice(choices)
user_choice = input(f'Enter your choice {choices}: ')


print(f'''
Computer choice: {comp_choice}
User choice: {user_choice}
''')

if user_choice in choices:

    if comp_choice == user_choice:
        print('Draw Tie!!🤝')
    
    elif comp_choice == 'rock':
        if user_choice == 'paper':
            print('User Won 💰<> Comp Lose 🥹!!')
        else:
            print('Comp Won 💰<> User Lose 🥹!!')
    
    elif comp_choice == 'paper':
        if user_choice == 'scissors':
            print('User Won 💰<> Comp Lose 🥹!!')
        else:
            print('Comp Won 💰<> User Lose 🥹!!')
    
    elif comp_choice == 'scissors':
        if user_choice == 'rock':
            print('User Won 💰<> Comp Lose 🥹!!')
        else:
            print('Comp Won 💰<> User Lose 🥹!!')
else:
    print('❌❌ Invalid Input or Wrong Choice!! ❌❌')
```
